```python
from PIL import Image
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
```

```python
os.chdir(r"C:\Users\hellozjf\AppData\Local\Temp\689fbcce1b24433c94e40884a05515f7")
questions = os.listdir()
X = np.zeros(shape=(0,4489))
y = np.arange(0)
for question in questions:
    print(question)
    os.chdir(question)
    cwd = os.getcwd()
    print(cwd)
    
    # 获取当前目录下面的所有图片
    pictures = os.listdir()
    for picture in pictures:
        print(cwd + "/" + picture)
        im = Image.open(picture)
        im = im.convert("L") 
        data = im.getdata()
        data = np.matrix(data)
        print(data.shape)
        X = np.vstack([X, data])
        y = np.hstack([y, question])
    
    os.chdir(r"..")
    
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=666)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(X_train, y_train)
```

```python
y_predict = knn_clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
